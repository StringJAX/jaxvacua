```python
from jax.typing import ArrayLike
from jax import Array
```

```python
def DW(x,cx,t,ct,fluxes,conj=True):
    
    imx = -(x-cx)*1j/2
    rex = (x+cx)/2
    imt = -(t-ct)*1j/2
    ret = (t+ct)/2
    
    #return jnp.array([rex[0]+1j*imx[0],rex[1]+1j*imx[1],ret+1j*imt])
    #return jnp.array([rex[0]**2,rex[1],ret])
    #return jnp.array([x[0]**2,x[1],t])
    if conj:
        return jnp.array([cx[0]**2,cx[1],ct])
    else:
        return jnp.array([x[0]**2,x[1],t])


def dDW_z(
    moduli: ArrayLike, moduli_c: ArrayLike, tau: complex, tau_c: complex, 
    fluxes: ArrayLike, conj: bool = False
    ) -> Array:

    return jax.jacfwd(DW,argnums=0,holomorphic=True)(moduli,moduli_c,tau,tau_c,fluxes,conj=conj)


def dDW_tau(
    moduli: ArrayLike, moduli_c: ArrayLike, tau: complex, tau_c: complex, 
    fluxes: ArrayLike, conj: bool = False
    ) -> Array:
    

    return jax.jacfwd(DW,argnums=2,holomorphic=True)(moduli,moduli_c,tau,tau_c,fluxes,conj=conj)

    
def dDW(
    moduli: ArrayLike, moduli_c: ArrayLike, tau: complex, tau_c: complex, 
    fluxes: ArrayLike, conj: bool = False
    ) -> Array:


    dDWz = dDW_z(moduli,moduli_c,tau,tau_c,fluxes,conj=conj)
    dDWtau = dDW_tau(moduli,moduli_c,tau,tau_c,fluxes,conj=conj)

    return jnp.append(dDWz,jnp.array([dDWtau]).T,axis=1)


def dDW_cz(
    moduli: ArrayLike, moduli_c: ArrayLike, tau: complex, tau_c: complex, 
    fluxes: ArrayLike, conj: bool = False
    ) -> Array:

    return jax.jacfwd(DW,argnums=1,holomorphic=True)(moduli,moduli_c,tau,tau_c,fluxes,conj=conj)


def dDW_ctau(
    moduli: ArrayLike, moduli_c: ArrayLike, tau: complex, tau_c: complex, 
    fluxes: ArrayLike, conj: bool = False
    ) -> Array:

    return jax.jacfwd(DW,argnums=3,holomorphic=True)(moduli,moduli_c,tau,tau_c,fluxes,conj=conj)


def dDW_c(
    moduli: ArrayLike, moduli_c: ArrayLike, tau: complex, tau_c: complex, 
    fluxes: ArrayLike, conj: bool = False
    ) -> Array:


    dDWz = dDW_cz(moduli,moduli_c,tau,tau_c,fluxes,conj=conj)
    dDWtau = dDW_ctau(moduli,moduli_c,tau,tau_c,fluxes,conj=conj)

    return jnp.append(dDWz,jnp.array([dDWtau]).T,axis=1)
```
